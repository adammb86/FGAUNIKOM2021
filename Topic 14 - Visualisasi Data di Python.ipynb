# Visualisasi Pie Chart

(Dokumentasi lengkap di https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.pie.html)

Pie chart cocok digunakan untuk menampilkan komposisi antar kategori yang bersifat statis terhadap jumlah keseluruhan data. Visualisasi ini akan optimal apabila jumlah kategori tidak terlalu banyak. Mari kita coba untuk membuat pie chart menggunakan data berikut:

<b>Data Penggemar Buah Tropis</b>

1. Nanas: 100 orang
2. Durian: 1000 orang
3. Rambutan: 300 orang
4. Manggis: 900 orang
5. Melon: 700 orang

```python
import matplotlib.pyplot as plt 

buah = ['Nanas', 'Durian', 'Rambutan', 'Manggis', 'Melon']
penggemar = [100, 1000, 300, 900, 700] 
plt.pie(penggemar, labels=buah) 

plt.show()
```

><b>Catatan:</b> Kita bisa menggunakan list, tuple, atau numpy array sebagai parameter yang mau dimunculkan pada pie chart

Kalau kita lihat bentuk visualisasi belum optimal dalam menunjukkan data. Sebagai contoh, porsi antara Durian dengan Manggis hampir serupa sehingga pembaca data akan kebingungan dalam menentukan mana yang lebih dominan. Kita bisa menambahkan parameter ``autopct`` (autopercent) untuk memberikan persentase kepada pice chart tersebut. Berikut kode programnya:

```python
plt.pie(
    penggemar, 
    labels=buah, 
    autopct= '%1.1f%%'
)

plt.show()
```

Kita juga bisa memberikan warna custom pada setiap elemen data di dalam pie chart agar lebih presentatif. Sebagi contoh kita berikan warna merah pada rambutan dan kuning pada Durian. Tambahkan parameter colors untuk mengkustom warna pada pie chart. Berikut contohnya:

```python
warna = ['#E8B824', '#EFD298', '#E54A48','#F2A2A9', '#758D20']

plt.pie(
    penggemar, 
    labels=buah, 
    autopct= '%1.1f%%',
    colors = warna
)

plt.show()
```

Apabila kita ingin mengekspos salah satu elemen data pada pie chart, kita bisa menggunakan parameter ``explode``. Sebagai contoh kita mau mengeskpos buah Manggis dalam visualisasi data yang sudah dibuat. Berikut kode programnya:

```python
buah_expose = [0, 0, 0, 0.2, 0]

plt.pie(
    penggemar, 
    labels=buah, 
    autopct= '%1.1f%%',
    colors = warna,
    explode = buah_expose
)

plt.show()
```

Sedikit sentuhan terakhir, kita bisa mempercantik pie chart dengan memberikan efek shadow agar terlihat tiga dimensi. Untuk melakukan hal tersebut, kita bisa menambahkan parameter ``shadow = True``. Berikut contoh penggunaannya:

```python
plt.pie(
    penggemar, 
    labels=buah, 
    autopct= '%1.1f%%',
    colors = warna,
    explode = buah_expose,
    shadow = True
)

plt.show()
```

Jangan lupa juga untuk memberikan judul pada diagram pie chart yang sudah dibuat agar pembacanya lebih jelas dalam menginterpretasikan maknanya. Berikut kode programnya:

```python
plt.title('Buah Tropis Favorit')

plt.pie(
    penggemar, 
    labels=buah, 
    autopct= '%1.1f%%',
    colors = warna,
    explode = buah_expose,
    shadow = True
)

plt.show()
```

# Diagram Batang

Visualisasi berikutnya adalah diagram batang. Diagram batang cocok digunakan untuk membandingkan data kategorical baik antar waktu maupun antar item data. Kategori yang dibandingkan jumlahnya bisa lebih banyak dibandingkan ketika kita menggunakan pie chart. Sebagai contoh mari kita visualisasikan data pada tabel berikut:

<b>Data Penduduk Jawa, Bali. NTT, dan NTB Tahun 2020</b>

| Propinsi | Jumlah Penduduk |
| :--- | :--- |
| Jawa Barat | 49.565.200 |
| Jawa Tengah | 34.738.200 |
| Jawa Timur | 39.955.900 |
| DKI Jakarta | 10.576.400 |
| DI Yogyakarta | 3.919.200 |
| Banten | 12.895.300 |
| Bali | 4.414.400 |
| Nusa Tenggara Timur | 5.513.400 |
| Nusa Tenggara Barat | 5.225.900 |

```python
import matplotlib.pyplot as plt
import numpy as np

propinsi = ['Jawa Barat', 'Jawa Tengah', 'Jawa Timur', 'DKI Jakarta', 'DI Yogyakarta', 'Banten', 'Bali', 'Nusa Tenggara Timur',  'Nusa Tenggara Barat']
jumlah_penduduk = [49565200, 34738200, 39955900, 10576400, 3919200, 12895300, 4414400, 5513400, 5225900]

x_coords = np.arange(len(propinsi))
plt.bar(
    x_coords, 
    jumlah_penduduk, 
    tick_label= propinsi
) 

plt.xticks(rotation=90) 
plt.ylabel('Jumlah Penduduk') 
plt.title('Data Penduduk Jawa, Bali. NTT, dan NTB Tahun 2020') 

plt.show()
```

Apabila ingin mengurutkan tampilan visualisasi di diagram batang, kita harus mengurutkan datanya secara manual menggunakan dataframe di pandas karena matplotlib tidak menyediakan fasilitas khusus kebutuhan ini. Berikut contoh kasusnya:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

propinsi = ['Jawa Barat', 'Jawa Tengah', 'Jawa Timur', 'DKI Jakarta', 'DI Yogyakarta', 'Banten', 'Bali', 'Nusa Tenggara Timur',  'Nusa Tenggara Barat']
jumlah_penduduk = [49565200, 34738200, 39955900, 10576400, 3919200, 12895300, 4414400, 5513400, 5225900]

data_penduduk = pd.DataFrame({'Propinsi': propinsi, 'Jumlah Penduduk': jumlah_penduduk}) 
data_penduduk.sort_values(by='Jumlah Penduduk', inplace=True)

x_coords = np.arange(len(propinsi))
plt.bar(
    x_coords, 
    data_penduduk['Jumlah Penduduk'], 
    tick_label= data_penduduk['Propinsi']
) 

plt.xticks(rotation=90) 
plt.ylabel('Jumlah Penduduk') 
plt.title('Data Penduduk Jawa, Bali. NTT, dan NTB Tahun 2020') 

plt.show()
```

Dengan tampilan berikut, kita akan lebih mudah dalam memahami data di dalam visualisasinya. Apabila kita ingin mengekspos salah satu elemen data pada diagram batang, kita bisa memanfaatkan parameter ``color``. Berikut contoh penggunaannya:

```python
warna = ['#0000FF', '#0000FF', '#0000FF', '#0000FF', '#0000FF', '#0000FF', '#0000FF', '#FF0000', '#0000FF']

x_coords = np.arange(len(propinsi))
plt.bar(
    x_coords, 
    data_penduduk['Jumlah Penduduk'], 
    tick_label= data_penduduk['Propinsi'],
    color = warna
) 

plt.xticks(rotation=90) 
plt.ylabel('Jumlah Penduduk') 
plt.title('Data Penduduk Jawa, Bali. NTT, dan NTB Tahun 2020') 

plt.show()
```

# Diagram Garis

Diagram garis baik digunakan untuk memvisualisasikan perbandingan suatu nilai (perkembangan) antar waktu. Diagram ini bisa digunakan untuk deret waktu yang panjang atau untuk membandingkan antar kategori dengan item yang terbatas. Sebagai contoh mari kita visualisasikan data berikut:

<b>Data Pencatatan Jumlah Pengunjung Restoran X di tanggal 22 Agustus 2021</b>

Asumsi: Data dicatat mulai dari jam 6 pagi sampai dengan jam 21. Pencatatan data dicatat sekali dalam 1 jam.

100, 102, 98, 70, 300, 305, 546, 690, 120, 145, 43, 57, 600, 987, 800, 234

Mari kita visualisasikan:

```python
jam_catat = [x for x in range(6,22)]
jumlah_pengunjung = [100, 102, 98, 70, 300, 305, 546, 690, 120, 145, 43, 57, 600, 987, 800, 234]

plt.plot(
    jam_catat, 
    jumlah_pengunjung
) 

plt.title('Data Jumlah Pengunjung Restoran X di tanggal 22 Agustus 2021') 
plt.ylabel('Jumlah Pengunjung') 
plt.xlabel('Jam Pencatatan') 
plt.show()
```

Kita juga bisa menambahkan marker untuk penanda poin. Berikut caranya:

```python
plt.plot(
    jam_catat, 
    jumlah_pengunjung,
    marker = 'x'
) 

plt.title('Data Jumlah Pengunjung Restoran X di tanggal 22 Agustus 2021') 
plt.ylabel('Jumlah Pengunjung') 
plt.xlabel('Jam Pencatatan') 
plt.show()
```

Selain dimanfaatkan untuk melihat perkembangan antar waktu, kita juga bisa menggunakan diagram garis untuk komparasi dengan kategori lain antar waktu. Sebagai contoh, kita mau membandingkan antara angka pengunjung aktual dengan yang kita prediksi berdasarkan hari-hari sebelumnya. Berikut contoh kasusnya:

```python
jumlah_pengunjung_prediksi = [90, 100, 98, 90, 296, 300, 500, 500, 127, 154, 50, 40, 700, 966, 845, 100]

plt.plot(
    jam_catat, 
    jumlah_pengunjung,
    marker = 'x',
    label = 'Aktual'
) 

plt.plot(
    jam_catat,
    jumlah_pengunjung_prediksi,
    linestyle = '--',
    marker = 'x',
    label = 'Prediksi'
)

plt.legend()
plt.title('Data Jumlah Pengunjung (Aktual dan Prediksi) Restoran X di tanggal 22 Agustus 2021') 
plt.ylabel('Jumlah Pengunjung') 
plt.xlabel('Jam Pencatatan') 
plt.show()
```

# Scatter Plot

Scatter plot merupakan bentuk visualisasi yang cocok digunakan untuk menggambarkan keterhubungan antar dua item atau distribusi antar dua item. Scatter plot juga bisa membantu kita dalam menemukan pencilan atau outlier dalam fase pemahaman data dalam ilmu sains data. Sebagai contoh kita mau melihat hubungan antara tinggi badan dan berat badan. Datanya adalah sebagai berikut:

| Nama Orang | Tinggi Badan | Berat Badan |
| :--- | :--- | :--- |
| A | 179 | 79 |
| B | 170 | 70 |
| C | 175 | 79 |
| D | 178 | 78 |
| E | 175 | 77 |
| F | 175 | 77 |
| G | 180 | 80 |
| H | 182 | 80 |
| I | 179 | 89 |
| J | 178 | 78 |

```python
nama = 'A B C D E F G H I J'.split()
tinggi = [179, 170, 175, 178, 175, 175, 180, 182, 179, 178]
berat = [79, 70, 79, 78, 77, 77, 80, 80, 89, 78]

plt.scatter(
    tinggi, 
    berat
) 

plt.title('Distribusi Berat dan Tingi Badan Peserta AKMIL') 
plt.ylabel('Berat Badan') 
plt.xlabel('Tinggi Badan') 

for i, name in enumerate(nama):
    plt.annotate(name, (tinggi[i], berat[i]))

plt.tight_layout()
plt.show()
```

Kita bisa melihat bahwa nama ``B`` dan `` I`` memiliki kombinasi berat dan tinggi badan yang sangat berbeda dengan orang lainnya. Bisa saja kita asumsikan di awal bahwa ``B`` dan ``I`` merupakan pencilan. Dibutuhkan metode atau pengolahan lainnya untuk memastikan pencilan akan tetapi dengan visualisasi ini kita mendapatkan informasi awal dengan cepat.

Kita juga bisa melihat distribusi dari dua item yang berbeda. Sebagai contoh kita ingin melihat data berat dan diameter dari jeruk nipis dan Lemon. Berikut contoh kasusnya:

```python
lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15, 9.96, 7.21, 10.04, 10.2, 11.06] 
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93, 132.93, 138.92, 145.98, 148.44, 152.81] 

lime_diameter = [6.15, 7.0, 7.0, 7.69, 7.95, 7.51, 10.46, 8.72, 9.53, 10.09] 
lime_weight = [112.76, 125.16, 131.36, 132.41, 138.08, 142.55, 156.86, 158.67, 163.28, 166.74] 

plt.scatter(
    lemon_diameter, 
    lemon_weight,
    color = 'y'
) 

plt.scatter(
    lime_diameter, 
    lime_weight,
    color = 'g'
)

plt.title('Lemon VS. Jeruk Nipis') 
plt.xlabel('Diameter (cm)') 
plt.ylabel('Berat (g)') 

plt.legend(['Lemons', 'Jeruk Nipis'], loc= 'upper left')
plt.show()
```

# Korelasi

Korelasi adalah pengukuran yang dilakukan untuk melihat sejauh mana nilai saling ketergantungan antar variabel-variabel data yang ada. Korelasi seringkali diartikan sebagai causation (sebab-akibat) padahal kedua hal itu berbeda. Korelasi tidak menggambarkan fungsi sebab-akibat dalam menyatakan ketergantungan antar variabelnya. Analisis korelasi biasanya digunakan sebagai bentuk data preprocessing ketika akan ekstraksi fitur (kolom data mana saja yang akan digunakan dalam pemodelan apabila bentuk data adalah data terstruktur). Metode yang paling populer dalam menentukan korelasi adalah metode Pearson Correlation.

Pearson menyatakan korelasi antar variabel dengan memberikan nilai antara -1 sampai dengan 1. Penjelasan rentang nilai dari Pearson Correlation dapat dilihat pada tabel berikut:

| Nilai Koefisien | Keterangan |
| :--- | :--- |
| Mendekati 1 | Total korelasi linear positif |
| 0 | Tidak ada korelasi linear. Kedua variabel kemungkinan besar tidak saling mempengaruhi|
| Mendekati -1 | Total korelasi linear negatif |

Dalam menghitung nilai korelasi, kita akan diminta menghitung juga nilai P. Nilai P menyatakan probabilitas bahwa antar kedua variabel yang ketergantungan signifikan secara statistik. Semakin kecil nilai P maka semakin tinggi dan semakin yakin bahwa antar item tersebut saling ketergantungan. Jika nilai p sebesar 0.05 (5%) maka 95% nilai korelasi tersebut signifikan. Sebaran nilai P dapat dilihat pada tabel berikut:

| Nilai P | Keterangan |
| :--- | :--- |
| < 0.001 | Bukti kuat bahwa korelasi signifikan |
| < 0.05 | Bukti moderat bahwa korelasi signifikan |
| < 0.1 | Bukti lemah bahwa korelasi signifikan |
| > 0.1 | Tidak ada bukti bahwa korelasi signifikan |

Pada topik 10, cara menghitung nilai korelasi antar atribut sudah dijelaskan. Di topik ini, mari kita fokus ke menghtiung nilai P dan memvisualisasikan korelasinya. Mari kita mulai dengan menghitung nilai P. Library yang digunakan untuk menghitung nilai P adalah ``scipy`` di sub module ``stats``. Sebagai contoh kita akan menghitung korelasi dan nilai P untuk atribut ``wheel-base`` dengan ``price``. Berikut kode programnya:

# Visualisasi Grouping

Seperti pada bahasan sebelumnya tentang pandas, kita bisa mengolah statistik berdasarkan kelompok kategori tertentu di suatu atribut atau lebih. Istilah itu dikenal sebagai grouping (group by). Mari kita lihat proses grouping pada data mobil dari file CSV.

Sebagai contoh, mari kita mencoba meng-grouping data mobil berdasarkan atribut ``drive-wheels``. Berikut kode programnya:

<b>Mengetahui kategori unique yang ada di atribut ``drive-wheels``</b>

```python
import pandas as pd
import numpy as np

df['drive-wheels'].unique()
```

Sebagai contoh, kita slicing saja dataframe ``df`` menjadi dataframe yang hanya berisikan ``drive-wheels``, ``body-style``, dan ``price``

<b>Slicing dataframe dan Menghitung rata-rata dari setiap kategori di atribut ``drive-wheels``</b>

```python
df_group_one = df[['drive-wheels','body-style','price']]
display(df_group_one)

df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean() 
display(df_group_one)
```

Kita juga bisa membuat group yang didasarkan pada dua atribut atau lebih. Berikut contohnya:

```python
df_gptest = df[['drive-wheels','body-style','price']] 
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean() 
grouped_test1
```

Setelah di-grouping, kita bisa melakukan pivot terhadap  data di ``grouped_test1``. Berikut kode programnya:

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style') 
grouped_pivot
```

Mari kita tangani missing value dengan mengisikan 0 ke missing value. Berikut kode programnya:

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0 
grouped_pivot
```

Kita bisa memvisualisasikan data grouping menggunakan heatmap. Berikut kode programnya:

```python
fig, ax = plt.subplots() 

im = ax.pcolor(grouped_pivot, cmap='RdBu') 

#label names 
row_labels = grouped_pivot.columns.levels[1] 
col_labels = grouped_pivot.index 

#move ticks and labels to the center 
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False) 
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False) 

#insert labels 
ax.set_xticklabels(row_labels, minor=False) 
ax.set_yticklabels(col_labels, minor=False) 

#rotate label if too long plt.xticks(rotation=90) 
fig.colorbar(im) 
plt.show()
```
