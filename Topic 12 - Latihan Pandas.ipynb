<b>Buat Dataframe baru dengan nama ``df_customer_favorite`` yang hanya berisi 5 data dari dataframe customer</b>

```python
df_customer_favorite = data_customer.head()

df_customer_favorite = data_customer.loc[11:15].copy()
display(df_customer_favorite)
```

<b>Tambahkan dua kolom pada ``df_customer_favorite`` bernama ``tanggallahir`` dan ``umur`` dimana kolom umur dihitung berdasarkan tanggal lahir</b>

```python
series_tanggal_lahir = pd.to_datetime(['2000-01-1', '1997-11-30', '2000-02-22', '1997-11-8', '1999-11-10'])
df_customer_favorite['tanggallahir'] = series_tanggal_lahir
display(df_customer_favorite)

hari_ini = dt.date.today().year
df_customer_favorite['umur'] =  hari_ini - df_customer_favorite['tanggallahir'].dt.year
display(df_customer_favorite)
```

# Latihan Dua

<b>Baca data dari file ``transactions_1k.csv`` ke dalam dataframe</b>

```python
data_transaksi = pd.read_csv('transactions_1k.csv')

display(data_transaksi)
```

<b>Munculkan data transaksi untuk CustomerID lebih dari 10000</b>

```python
seleksi = data_transaksi['CustomerID'] > 10000

display(data_transaksi[seleksi])
```

<b>Munculkan data transaksi yang terjadi pada tanggal 23 Agustus 2012 sampai dengan 24 Agustus 2012</b>

```python
seleksi = (data_transaksi['Date'] >= '2012-08-23') & (data_transaksi['Date'] <= '2012-08-24')
display(data_transaksi[seleksi])
```

<b>Munculkan data transaksi dengan jumlah pembelian terbanyak</b>

```python
seleksi = data_transaksi['Amount'] == data_transaksi['Amount'].max()

display(data_transaksi[seleksi])
```

<b>Munculkan data transaksi dengan total pembayaran terkecil</b>

```python
seleksi = data_transaksi['Price'] == data_transaksi['Price'].min()

display(data_transaksi[seleksi])
```

<b>Gunakan fungsi ``describe`` untuk menampilkan pengolahan statistik untuk semua kolom</b>

```python
data_transaksi.describe(include= 'all')
```

<b>Tambahkan kolom tanggal, bulan, tahun, dan jam yang merupakan hasil dari splitting kolom ``Date`` dan ``Time``</b>

```python
data_transaksi['tanggal'] = pd.to_datetime(data_transaksi['Date']).dt.day
data_transaksi['bulan'] = pd.to_datetime(data_transaksi['Date']).dt.month
data_transaksi['tahun'] = pd.to_datetime(data_transaksi['Date']).dt.year
data_transaksi['jam'] = pd.to_datetime(data_transaksi['Time']).dt.hour

display(data_transaksi)
```

<b>Munculkan list Top 5 Customers yang memiliki nilai transaksi paling banyak</b>

```python
group_customer = data_transaksi.groupby(by = 'CustomerID', as_index= False)
sum_of_price = group_customer['Price'].sum()
sort_price = sum_of_price.sort_values(by='Price', ascending= False)

display(sort_price.head())
```

<b>Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi !</b>

```python
groupby_gas_station = data_transaksi.groupby('GasStationID')
sum_of_price = groupby_gas_station['Price'].sum().to_frame()

display(sum_of_price)
```

```python
groupby_gas_station = data_transaksi.groupby(['tanggal','jam'])
sum_transaction = groupby_gas_station['Price'].sum().to_frame()

display(sum_transaction)
```

<b>Muculkan 5 nama customer yang memiliki total nilai transaksi terbesar</b>

```python
group_transaction = data_transaksi.groupby(by= 'CustomerID', as_index= False)
sum_of_price = group_transaction['Price'].sum()
sort_price = sum_of_price.sort_values(by= 'Price', ascending=False)

display(sort_price)

data_customer_lengkap = pd.merge(sort_price,data_customer,how='inner',on='CustomerID')
display(data_customer_lengkap[['Segment','Price']].head())
```
